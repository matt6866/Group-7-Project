```python
Days_On_Market_all = pd.read_csv('DaysOnMarket.csv')
Days_On_Market= Days_On_Market_all.dropna()
median_rent_all= pd.read_csv('MedianAskingRentByRoomType.csv')
median_rent= median_rent_all.dropna()
```

```python
alt.data_transformers.disable_max_rows()
interval = alt.selection_interval()
type(interval)
```

```python
circle_chart = alt.Chart(Days_On_Market)
circle_chart = circle_chart.mark_circle()

circle_chart = circle_chart.encode(
    x='DaysOnMarket',
    y='Borough')
circle_chart = circle_chart.encode(    
    color=alt.condition(predicate=interval, if_true='Borough', if_false=alt.value('blue')))

circle_chart = circle_chart.properties(
    width=700,
    height=400,
    selection=interval,
    title = "MI Days On The Market Circle Chart"
)

type(circle_chart)
circle_chart
```

```python
day_on_market_scales = alt.selection_interval(bind='scales')

alt.Chart(Days_On_Market).mark_point().encode(
    y='DaysOnMarket:Q',
    x='Date:T',
    color='Borough:N'
).properties(
    title = "MI Days On The Market Scales",
).add_selection(
    day_on_market_scales
)
```

```python
source = Days_On_Market
highlight = alt.selection(type='single', on='mouseover',
                          fields=['symbol'], nearest=True)
base = alt.Chart(source).encode(
    x='Date:T',
    y='DaysOnMarket:Q',
    color='Borough:N')
points_days = base.mark_circle().encode(
    opacity=alt.value(0)
).add_selection(
    highlight
).properties(
    width=600,
    height = 500,
    title ="MI Days On The Market - Lines and Points"
)
lines_days = base.mark_line().encode(
    size=alt.condition(~highlight, alt.value(1), alt.value(3))
)
points_days + lines_days
```
